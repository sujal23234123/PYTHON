```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[6,5,4,3,2]

plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[5,6,7,8]

plt.plot(x,y,color="red",linestyle="--",marker="o",linewidth="3",markersize=10)
plt.grid(True)
plt.show()
```


Multiple plots 

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y1=[5,7,19,8,2]
y2=[1,2,3,4,5]

plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
plt.plot(x,y1,color="red")

plt.subplot(2,2,2)
plt.plot(x,y2,color="green")
plt.title=("Plot 2")


plt.subplot(2,2,3)
plt.plot(y2,y1,color="blue")
plt.title=("Plot 3")

plt.subplot(2,2,4)
plt.plot(y1,y2,color="black")
```

Types of plots


```python
#Bar plot

import matplotlib.pyplot as plt

x=[2,5,3,8,1]
y=[7,6,9,8,5]

plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar plot")
```

```python
#histograms
import matplotlib.pyplot as plt
x=[3,4,5,6,7,5,3,3,2,3,4,5,2]


plt.hist(x,bins=6,color="purple",edgecolor="black")
```

```python
#scatter plot

x=[4,5,4,7,6]
y=[8,6,4,3,3]

plt.scatter(x,y,color="red",marker="x")
```

```python
#pie chart

labels=["A","B","C","D"]
size=[10,20,30,40]
color=["green","red","yellow","blue"]
explode=(0,0,0.3,0)

plt.pie(size,labels=labels,autopct="%1.1f%%",shadow=True,explode=explode)
```
