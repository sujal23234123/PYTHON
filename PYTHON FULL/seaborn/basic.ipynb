```python
import seaborn as sns

tips=sns.load_dataset("tips")            #load_dataset are the load data in seaborn for practice datasets
tips
```

Differents plots

```python
#scatter plot
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x="tip",y="size",data=tips)
plt.title(" scatter plot of tips vs size plot")
plt.xlabel("tip")
plt.ylabel("size")

```

```python

tips.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(x="day",y="time",data=tips)
plt.title("line plot of day vs time")
plt.xlabel("day")
plt.ylabel("time")
```

```python
tips.info()
```

```python
#categorical plot(barplot)

import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x="day",y="total_bill",data=tips)
plt.title("barplot of day vs total bill")
plt.xlabel("day")
plt.ylabel("total bill")
```

```python
#box plot

import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x="total_bill",y="smoker",data=tips)
plt.title("box plot of total bill vs smoker")
plt.xlabel("total bill")
plt.ylabel("smoker")
```

```python
#violin plot

import matplotlib.pyplot as plt
import seaborn as sns

sns.violinplot(x="day",y="tip",data=tips)
plt.title("violin plot of day vs tip")
plt.xlabel("day")
plt.ylabel("tip")
```

```python
#histogram

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(tips["total_bill"],bins=10,edgecolor="black",kde=True)
plt.title("hist plot of total bill")
```

```python
#pairplot

import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(tips)
```

```python
#correlation and that heatmap 

corr=tips[["total_bill","size","tip"]].corr()
corr

sns.heatmap(corr,annot=True,cmap="coolwarm")
```
