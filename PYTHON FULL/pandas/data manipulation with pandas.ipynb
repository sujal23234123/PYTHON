```python
import pandas as pd

df=pd.read_csv("data.csv")
df.head(5)
df.drop(15,inplace=True)
df.head(5)
```

```python
#handling missing value in data

df.isnull().any()
df.isnull().sum()

#add item to that missing value
df.fillna(0).head(5)
```

```python
#rename coloumn name

df=df.rename(columns={"Date":"sales date"})
df.head(5)
```

```python
#update a data and make a new coloumn

df["new value"]=df["Value"].apply(lambda x:x*2)
df.head(5)
```

```python
#aggregating and grouping
grouped_data=df.groupby(["Product"])["Sales"].sum()
grouped_data.head(5)
```

```python
doublegroup=df.groupby(["Product","Sales"])["Value"].sum()
doublegroup
```

```python
df.head(5)
```

```python
#applying aggregrated function at one time

multipleagg=df.groupby(["Category","Value"])["Sales"].agg(["mean","sum","count"])
multipleagg.head(5)
```

```python
#merging and joining dataframe

import pandas as pd

df1=pd.DataFrame({"key":["A","B","C"],"value":["1","2","3"]})
df2=pd.DataFrame({"key":["A","B","D"],"value":["4","5","6"]})

```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
